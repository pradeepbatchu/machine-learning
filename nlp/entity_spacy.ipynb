```python
import wikipedia
import spacy
import textacy
import en_core_web_sm

subject = 'Bill Gates'

#The Wikipedia Page
wikiResults = wikipedia.search(subject)
print("wikiResults:", wikiResults)
wikiPage = wikipedia.page(wikiResults[1]).content
#wikiPage = wikipedia.page('Bill Gates').content
print("\n\nwikiPage:", wikiPage, "'\n")
#nlp = en_core_web_sm.load()
nlp = spacy.load('en_core_web_lg')
```

```python
import re
import nltk
import string
 # Remove URLs
text = re.sub(r"http\S+", "", wikiPage)
    
    # Remove emails
text = re.sub(r'[\w\.-]+@[\w\.-]+', ' ', text, flags=re.MULTILINE)
```

```python
document = nlp(text)
#uniqueStatements = set()
```

```python
# 'doc' now contains a parsed version of text. We can use it to do anything we want!
# For example, this will print out all the named entities that were detected:
for entity in document.ents:
    print(f"{entity.text} ({entity.label_})")
```
